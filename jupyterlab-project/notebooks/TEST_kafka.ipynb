```python
from kafka import KafkaAdminClient, KafkaProducer
from kafka.admin import NewTopic
from kafka.errors import TopicAlreadyExistsError

admin = KafkaAdminClient(bootstrap_servers="broker:9092", client_id='setup-client')
topic_name = "test-topic"

# Sprawdzenie i tworzenie topicu (jeśli nie istnieje)
existing_topics = admin.list_topics()

if topic_name in existing_topics:
    print(f"✅ Topic '{topic_name}' już istnieje - pomijam tworzenie")
else:
    print(f"Tworzę topic '{topic_name}'...")
    topic_list = [NewTopic(name=topic_name, num_partitions=1, replication_factor=1)]
    try:
        admin.create_topics(new_topics=topic_list, validate_only=False)
        print(f"✅ Topic '{topic_name}' został pomyślnie utworzony")
    except TopicAlreadyExistsError:
        print(f"✅ Topic '{topic_name}' już istnieje")

print("\nAktualne tematy:", admin.list_topics())

# === Poprawiony test producenta ===
producer = KafkaProducer(bootstrap_servers="broker:9092")

# Poprawny sposób - wiadomość musi być typu bytes
message = b"Testowa wiad!"
future = producer.send(topic_name, value=message)
producer.flush()

print(f"✅ Wiadomość wysłana pomyślnie do topicu '{topic_name}'")
print("Kafka działa prawidłowo! 🎉")
```
